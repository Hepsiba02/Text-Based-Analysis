**RUN THE PIPELINE:**

```python
import pandas as pd
import regex as re

# Define the pipeline
pipeline = [str.lower, tokenize, remove_stop]

# Define the tokenize function
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)

# Define the remove_stop function
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]

# Define the prepare function
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens

# Specify the file path
file_path = "/Users/hepsiba/Downloads/archive (2)/Tweets.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

df['tokens'] = df['text'].apply(lambda x: prepare(str(x), pipeline=pipeline))

# Display the DataFrame with the processed tokens
print(df[['text', 'tokens']].head())
```

**COUNTING THE TOKENS:**

```python
#count the tokens
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

**WORD FREQUENCY ANALYSIS:**

```python
#Word frequency Analysis

from collections import Counter

tokens = tokenize("She likes my cats and my cats like my sofa.")

testcount= Counter(tokens)
print(testcount)
```

**ADDING DATA TO COUNT:**

```python
#Adding new data to the current count

more_tokens = tokenize("She likes dogs and cats.")
testcount.update(more_tokens)
print(testcount)
```

**COUNTNG:**

```python
#Let’s count

counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

**FREQUENCY OF DATAFRAME:**

```python
#Now, we put the word frequency in it’s own dataframe for analysis

freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
print(freq_df)
```

**SORT & DISPLAY**

```python
#Lets sort and display

freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

**WORDCLOUD:**

```python
from wordcloud import WordCloud

def create_wordcloud(negative_tweets):
    all_text = ' '.join(negative_tweets['negativereason'])
    wordcloud = WordCloud(width=800, height=400, background_color='black').generate(all_text)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Example for American Airlines:
negative_tweets = df.loc[(df['airline'] == 'American') & (df['airline_sentiment'] == 'negative')]
create_wordcloud(negative_tweets)
```
