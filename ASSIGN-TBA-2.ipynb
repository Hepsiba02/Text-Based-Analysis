**FUNCTION TO TOKENIZE:**

```python
import regex as re
def tokenize(text):
# Use the findall function from the regex module to find all matches of the pattern in the text
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

```python
# Assuming stopwords are typical English language stopwords, define them as follows:
stopwords = set([
    "i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours",
    "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers",
    "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves",
    "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are",
    "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does",
    "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until",
    "while", "of", "at", "by", "for", "with", "about", "against", "between", "into",
    "through", "during", "before", "after", "above", "below", "to", "from", "up", "down",
    "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here",
    "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more",
    "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so",
    "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"
])

def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]

# Test the function
testtokens = ['This', 'is', 'a', 'test', 'of', 'the', 'function', 'with', 'stopwords']
cleantokens = remove_stop(testtokens)
print(cleantokens)
```

```python
from nltk.corpus import stopwords

# Existing stopwords
stop_words = set(stopwords.words('english'))

# Add a custom word to the stopword list
stop_words.add('sample')

# Remove a word from the stopword list
stop_words.discard('is')

print(stop_words)
```

```python
pipeline = [str.lower, tokenize, remove_stop]

def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

**WORDCLOUD:**

```python
from wordcloud import WordCloud

def create_wordcloud(negative_tweets):
    all_text = ' '.join(negative_tweets['negativereason'])
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Example for American Airlines:
negative_tweets = df.loc[(df['airline'] == 'American') & (df['airline_sentiment'] == 'negative')]
create_wordcloud(negative_tweets)
```

```python
negative_tweets = df.loc[(df['airline'] == 'Southwest') & (df['airline_sentiment'] == 'negative')]
create_wordcloud(negative_tweets)
```
