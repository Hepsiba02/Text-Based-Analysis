**FUNCTION TO TOKENIZE:**

```python
import regex as re
def tokenize(text):
# Use the findall function from the regex module to find all matches of the pattern in the text
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
```

**TOKENIZING THE TEST TEXT:**

```python
# Defining a test text string
testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"
testtokens = tokenize(testtext)
# Printing the tokens extracted from the test text
print(testtokens)
```

**RETRIEVING THE SET OF ENGLISH STOPWORDS:** 

```python
# Importing the Natural Language Toolkit (nltk) library
import nltk 
# Downloading the stopwords corpus from nltk
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

**TO REMOVE STOPWORDS:**

```python
def remove_stop(tokens):
# Use list comprehension to iterate over each token in the input list,
# keeping only those tokens that are not stopwords (ignoring case)
    return [t for t in tokens if t.lower() not in stopwords]
#Just testing it
cleantokens= remove_stop(testtokens)
print(cleantokens)
```

**CREATING THE PIPELINE:**

```python
# Define a list called pipeline that contains functions to be applied sequentially
pipeline = [str.lower, tokenize, remove_stop]

def prepare(text, pipeline):
# Initialize tokens with the input text
    tokens = text
# Iterate over each function in the pipeline and apply it sequentially to the tokens
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

**RUN THE PIPELINE:**

```python
# Apply the prepare function to each row in the 'text' column of the DataFrame df
# using the specified pipeline of text processing functions
df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
```

**COUNTING THE TOKENS:**

```python
# Create a new column 'num_tokens' in the DataFrame df
# by mapping the len function to each element in the 'tokens' column
df['num_tokens'] = df['tokens'].map(len)
# Display the last few rows of the DataFrame df to inspect the changes
df.tail()
```

**WORD FREQUENCY ANALYSIS:**

```python
# Import the Counter class from the collections module
from collections import Counter

tokens = tokenize("She likes my cats and my cats like my sofa.")

# Create a Counter object to count the frequency of each token
testcount= Counter(tokens)
print(testcount)
```

**ADDING DATA TO COUNT:**

```python
# Tokenize the additional text "She likes dogs and cats." using the tokenize function
more_tokens = tokenize("She likes dogs and cats.")
# Update the existing Counter object testcount with the counts of tokens from the additional text
testcount.update(more_tokens)
print(testcount)
```

**COUNTING:** 

```python
counter = Counter()
# Update the Counter object with the tokens from each row of the 'tokens' column in the DataFrame df
# by mapping the update method to each element in the 'tokens' column
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

**FREQUENCY OF DATAFRAME:**

```python
#Now, we put the word frequency in it’s own dataframe for analysis
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

**SORT AND DISPLAY:**

```python
#Lets sort and display
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

**FINDING INFORMATION OF FREQUENCY:**

```python
# Generate descriptive statistics of the frequency DataFrame freq_df
freq_df.describe()
```

**BAR CHART:**

```python
# Create a horizontal bar plot of the top 15 tokens and their frequencies from the frequency DataFrame freq_df
# Set the width of the bars to 0.95 and the size of the figure to 12x8 inches
ax = freq_df.head(15).plot(kind='barh', width=0.95, figsize=(12,8))
ax.invert_yaxis()
ax.set(xlabel='Frequency', ylabel='Token', title='Top Words')
```
